## CODE RETRIEVAL FROM CSV FILE

```python
from langchain_community.document_loaders.csv_loader import CSVLoader

loader = CSVLoader(file_path="codes.csv")
docs  = loader.load()
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview")
```

```python
from langchain_core.vectorstores import InMemoryVectorStore

vector_store = InMemoryVectorStore(embeddings)
```

```python
document_ids = vector_store.add_documents(documents=docs)

print(document_ids[:3])
```

```python
context = vector_store.similarity_search("code for Plasma fresh frozen", k=4)
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
context
```

```python
cntx = ''
for i in context:
    cntx += i.page_content
    cntx += '\n'
    
```

```python
print(cntx)
```

```python
from langchain.messages import SystemMessage, HumanMessage

sys_msg = SystemMessage(content="Answer only from context.")
hmn_msg = HumanMessage(content=f"""question: code for Plasma fresh frozen?
                       
                       context:{cntx}""")


model.invoke([sys_msg, hmn_msg])
```
