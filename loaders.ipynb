PDF LOADER

```python
from langchain_community.document_loaders import PyPDFLoader

file_path = "wastage_of_plastic.pdf"
loader = PyPDFLoader(file_path)
loader.load()
```

WEBPAGE LOADER

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.geeksforgeeks.org/nlp/what-is-retrieval-augmented-generation-rag/")
loader.load()
```

CSV LOADER

```python
from langchain_community.document_loaders.csv_loader import CSVLoader

loader = CSVLoader(file_path="HCPCS Codes.csv")
loader.load()
```

JSON LOADER

```python
from langchain_community.document_loaders import JSONLoader

loader = JSONLoader(
    file_path="sample_json.json",
    jq_schema=".student",
    text_content=False,
)
loader.load()
```

HTML LOADER

```python
from langchain_community.document_loaders import BSHTMLLoader

loader = BSHTMLLoader(
    file_path="sample_html.html",
)
loader.load()
```
