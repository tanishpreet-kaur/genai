```python
document = """Tanishpreet is a Data scientist at ids. her email is tanishpreet.k@idsil.com"""
```

```python
from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage, SystemMessage

sys_msg = SystemMessage("tell me the name of the person and what he/she does (work)")
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
hum_msg = HumanMessage(document)
```

```python
from pydantic import BaseModel, field_validator

class PersonInfoModel(BaseModel):
    name: str
    work: str
    email: str
      
    @field_validator("email")
    @classmethod
    def check_email(cls, v):
        if v.endswith("idsil.com"):
            return v
        else:
            raise ValueError("Email domain must be idsil.com")
```

```python
structured_model = model.with_structured_output(PersonInfoModel)
```

```python

ans1 =  structured_model.invoke([sys_msg, hum_msg])
```
