```python
from langchain_core.tools import tool


@tool
def create_file(name: str):
    """this tool is used for create a file"""
    with open(name, "w") as f:
        f.write("hi! how are you?")
```

```python
create_file.invoke({"name":"sample_text.txt"})
```

```python
from langchain.agents import create_agent
from langchain.chat_models import init_chat_model
model = init_chat_model("google_genai:gemini-2.5-flash-lite")

agent = create_agent(model, tools=[create_file])
```

```python
ans = agent.invoke({"messages": [("user", "make a file named file-2.txt")]})

print(ans["messages"][-1].content)
```
