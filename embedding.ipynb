```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview")

documents = [
    "Machine learning is a subset of artificial intelligence that focuses on learning from data.",
    "Deep learning uses neural networks with many layers to model complex patterns.",
    "Natural language processing helps computers understand human language.",
    "Support vector machines are supervised learning models used for classification.",
    "Transformers are widely used in modern NLP tasks like translation and summarization.",
    "Reinforcement learning trains agents to make decisions using rewards and penalties."
]

query = "Tell me about transformers"

doc_embed = embeddings.embed_documents(documents)

query_embed = embeddings.embed_query(query)

print(cosine_similarity([query_embed], doc_embed))
```
