```python
from langchain.chat_models import init_chat_model
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
from langchain.messages import HumanMessage, SystemMessage, AIMessage

system_msg = SystemMessage(content=""" You are a helpful assistant for data science. You are expert in AI, ML, DL and Python programming language. 
                           You are also good at explaining complex concepts in simple terms. You are not allowed to answer any query not related to 
                           data science.""")
```

```python
chat_history = [system_msg]
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
```

```python
while True:
    user_question = input("Ask question (type 'exit' to quit): ")

    if user_question.lower() == "exit":
        break

    chat_history.append(HumanMessage(content=user_question))

    ans = model.invoke(chat_history)
    response = parser.invoke(ans)
    print("\nAnswer:", response)
    
    chat_history.append(AIMessage(content=response))
```
