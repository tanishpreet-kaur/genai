```python
from langchain.chat_models import init_chat_model
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
```

IMAGE INPUT

```python
# From URL
message = {
    "role": "user",
    "content": [
        {"type": "text", "text": "Describe the content of this image."},
        {"type": "image", "url": "https://miro.medium.com/v2/resize:fit:720/format:webp/1*J7vyY3EjY46AlduMvr9FbQ.png"},
    ]
}

img1 = model.invoke([message])
parser.invoke(img1)
```

```python
import base64
with open("image.png", "rb") as f:
    img_base64 = base64.b64encode(f.read()).decode("utf-8")
```

```python
# From base64 data
message = {
    "role": "user",
    "content": [
        {"type": "text", "text": "Describe the content of this image."},
        {
            "type": "image",
            "base64": img_base64,
            "mime_type": "image/png",
        },
    ]
}

img2 = model.invoke([message])
parser.invoke(img2)
```
