````python
from langchain.chat_models import init_chat_model
model = init_chat_model("google_genai:gemini-2.5-flash-lite", temperature=2)
````

````python
prompt = """

do sentiment analyses

```text
I really enjoyed this movie. The story was engaging, the acting felt natural, and the visuals were beautiful. It kept me interested from start to finish and left a good impression. Definitely worth watching! 🎬😊
```
give answer in one word

"""

ans = model.invoke("DOG")
````

````python
ans
````

````python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parsed_ans = parser.invoke(ans)
````

````python
print(parsed_ans)
````

````python
print(parsed_ans)   
````

````python
if parsed_ans=="Positive" or "+ve" :
    print("The sentiment of the text is positive.")
else:
    print("The sentiment of the text is negative.")
````

````python
from enum import Enum

class Sentiments(str, Enum):
    POSITIVE = "Positive"
    NEGATIVE = "Negative"
    NEUTRAL = "Neutral"
    
from typing import TypedDict
class Sentiment(TypedDict):
    sentiment: Sentiments
````

````python
model_with_structure = model.with_structured_output(Sentiment)
````

````python
ans2 = model_with_structure.invoke(prompt)
````

````python
ans2["sentiment"]
````

````python
if ans2["sentiment"] == Sentiments.POSITIVE:
    print("The sentiment of the text is positive.")
elif ans2["sentiment"] == Sentiments.NEGATIVE:
    print("The sentiment of the text is negative.")
else:
    print("The sentiment of the text is neutral.")
````

````python
from langchain.messages import HumanMessage, SystemMessage, AIMessage

system_msg = SystemMessage(content="You are a helpful assistant for data science. You are expert in AI, ML, DL and Python programming language. You are also good at explaining complex concepts in simple terms. You are not allowed to answer any query not related to data science.")
human_msg = HumanMessage(content="what is machine learning?")
````

````python
ans4 = model.invoke([system_msg, human_msg])
print(parsed_ans4:=parser.invoke(ans4))
````

````python
parsed_ans4
````

````python
ans4 = model.invoke([system_msg, human_msg, AIMessage(content=parsed_ans4), HumanMessage(content="tell me its types in detail")])
print(parser.invoke(ans4))
````
